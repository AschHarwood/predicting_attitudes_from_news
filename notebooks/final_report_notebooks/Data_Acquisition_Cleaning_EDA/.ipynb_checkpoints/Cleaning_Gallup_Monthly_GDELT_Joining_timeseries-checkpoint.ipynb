## Gallup Processing

```python
#reading in gallup
gallup = pd.read_csv('/floyd/home/Capstone/cap_notebooks/Gallup/Satisifaction_with_US - Sheet1.csv')
```

```python
#sort by date
gallup.sort_values('Date', inplace=True)
```

```python
#converting date to datetime object
gallup['Date'] = pd.to_datetime(gallup['Date'])
```

```python
#sort by date
gallup.sort_values('Date')
```

```python
#gallup.groupby(['Date']).mean()[3:]
```

```python
gallup.sort_values('Date', inplace=True)
```

```python
#dropping first three rows because out of text date range
gallup_short = gallup[3:]
```

```python
#reset index
gallup_short.reset_index(drop=True, inplace=True)
```

```python
#reviewing data
gallup_short
```

```python
#dropping rows 
gallup_short = gallup_short.drop([51, 52, 53, 54, 55])
```

```python
gallup_short.drop([20, 44], inplace=True)
```

```python
#merging text and gallup
tokens_target = pd.concat([tokens.reset_index(), gallup_short.reset_index()], axis=1)
```
